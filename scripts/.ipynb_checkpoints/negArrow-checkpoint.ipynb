```python
from modTool import alignScore, modPredict
import pysam
```

```python
samfile_neg = pysam.AlignmentFile("/scratch/gabai/addseq/data/ctrl/unique.0.pass.sorted.bam", "rb")
parse_bam_neg = {}
for line in samfile_neg:
    read = str(line)
    readID = read.split('\t')[0]
    strandID = int(read.split('\t')[1])
    if strandID == 0:
        strand = 1
    elif strandID == 16:
        strand = -1
    else:
        strand = 0
    parse_bam_neg[readID] = strand
samfile_neg.close()

eventAlign_neg = '/gicephfs/brookslab/bsaintjo/220516_ang_conc_unique/unique.0.eventalign.tsv'
```

```python
outputFile = '/scratch/gabai/addseq/data/ctrl/unique.0.arrow'
cawlrRead = alignScore(eventAlign = eventAlign_neg, downReads=500, downScore=5000, parse_bam = parse_bam_neg)
scored = {"scored": cawlrRead}
new_fdict = scored
output = Table.from_pydict(new_fdict, schema=table.schema)
feather.write_feather(output, outputFile)
print('done')
```
