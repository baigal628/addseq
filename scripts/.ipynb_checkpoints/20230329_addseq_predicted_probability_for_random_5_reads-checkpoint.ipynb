```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import torch
from torch.utils.data import DataLoader
from torchsummary import summary
from tqdm import tqdm
from nanopore_dataset import create_sample_map
from nanopore_dataset import create_splits
from nanopore_dataset import load_csv
from nanopore_dataset import NanoporeDataset

from resnet1d import ResNet1D
import seaborn as sns
```

```python
device = "cuda" if torch.cuda.is_available() else "mps" if torch.backends.mps.is_available() else "cpu"
```

```python
model = ResNet1D(
            in_channels=1,
            base_filters=128,
            kernel_size=3,
            stride=2,
            groups=1,
            n_block=8,
            n_classes=2,
            downsample_gap=2,
            increasefilter_gap=4,
            use_do=False)
summary(model, (1, 400), device= device)
```

```python
MESMLR_FN = './nanopore_classification/best_models/mesmlr_resnet1d.pt'
ADDSEQ_FN = './nanopore_classification/best_models/addseq_resnet1d.pt'

weights_path = ADDSEQ_FN
model.load_state_dict(torch.load(weights_path, map_location=torch.device(device)))
model.to(device)
model.eval()
```

```python
data_ctr = './nanopore_classification/data/addseq/reprocessed-unique.0.eventalign.signal.csv'
data_tr = './nanopore_classification/data/addseq/reprocessed-unique.500.eventalign.signal.csv'
addseq_neg = load_csv(data_ctr, min_val=50, max_val=130, max_sequences=10)
addseq_pos = load_csv(data_tr, min_val=50, max_val=130, max_sequences=10)
```

```python
input_len = 400
# n_predictions = len(seq_neg)-input_len+1
n_predictions = 10

for read in range(5):
    
    sequence_tensor_neg = torch.tensor(addseq_neg[read])
    sequence_tensor_pos = torch.tensor(addseq_pos[read])
    
    for i in tqdm(range(n_predictions)):
        xshift = i*100
        jump = 200
        start = i*jump
        input_tensor_neg[:, :, :] = sequence_tensor_neg[start:start+input_len]
        prediction_neg = model(input_tensor_neg).sigmoid().item()
        seq = input_tensor_neg.reshape(-1,)
        print('neg mean: ', torch.mean(seq))
        plt.plot(seq-torch.mean(seq).item() + xshift , torch.tensor(range(400)), marker='o', markeredgewidth=0, 
                 label = "neg", markersize=1.5, linewidth=1, color = 'blue')
        plt.text(xshift , 410, str(round(prediction_neg, 2)), rotation = 0, ha = 'center', color = 'blue')

        input_tensor_pos[:, :, :] = sequence_tensor_pos[start:start+input_len]
        prediction_pos = model(input_tensor_pos).sigmoid().item()
        seq = input_tensor_pos.reshape(-1,)
        print('pos mean: ', torch.mean(seq))
        plt.plot(seq-torch.mean(seq).item() + xshift , torch.tensor(range(400)), marker='o', markeredgewidth=0, 
             label = "pos", markersize=1.5, linewidth=1, color = 'orange')
        plt.text(xshift + 5 , 450, str(round(prediction_pos, 2)), rotation = 0, ha = 'center', color = 'orange')

    plt.tick_params(bottom=False, labelbottom=True,\
                   left=True, labelleft=False, \
                   right=False, labelright=True,\
                   top=False, labeltop=False)
#     plt.legend()
    outfile = '../results/figures/read_' + str(read) + '_pos_neg_signal.pdf'
    plt.show()
#     plt.savefig(outfile, dpi = 600)
    plt.close()
```

```python
for read in range(5):
    sequence_tensor_pos = torch.tensor(addseq_neg[read])
    
    for i in tqdm(range(n_predictions)):
        xshift = i*600
        jump = 200
        start = i*jump
        input_tensor_pos[:, :, :] = sequence_tensor_pos[start:start+input_len]
        seq = input_tensor_pos.reshape(-1,)
        prediction = model(input_tensor_pos).sigmoid().item()
        
        plt.plot(seq-torch.mean(seq).item() + xshift , torch.tensor(range(400)), marker='o', markeredgewidth=0, 
                     label = "pos", markersize=1.5, linewidth=1, color = 'C0')
        plt.text(xshift , 420, str(round(prediction, 2)), rotation = 90, va = 'bottom', ha = 'center', color = 'C0',  fontsize=8)

        for j in range(3):
            if j == 0:
                # random
                new_seq = input_tensor_pos[0, 0, torch.randperm(400)].reshape(1,1,400)
                col = 'C1'
                label = 'random'
            elif j == 1:
                # ascending
                new_seq = torch.sort(input_tensor_pos)[0]
                col = 'C2'
                label = 'ascend'
            else:
                #descending
                new_seq = torch.sort(input_tensor_pos, descending = True)[0]
                col = 'C3'
                label = 'descend'
            newshift = (j+1)*120
            seq = new_seq.reshape(-1,)
            new_prediction = model(new_seq).sigmoid().item()
            plt.plot(seq-torch.mean(seq).item() + xshift + newshift , torch.tensor(range(400)), marker='o', markeredgewidth=0, 
                     label = label, markersize=1.5, linewidth=1, color = col)
            plt.text(xshift + newshift , 420, str(round(new_prediction, 2)), rotation = 90, va = 'bottom',
                     ha = 'center', color = col, fontsize=8)
    
    plt.legend(['pos', 'random', 'ascend', 'descend'], loc = 'lower right')
#     plt.show()
    outfile = '../results/figures/read_' + str(read) + '_pos_signal_ordered.pdf'
    plt.savefig(outfile)
    plt.close()
```

```python
predict = []
count =0
for (seq_neg, seq_pos) in seq:
    print("neg signals:", seq_neg[:10])
    sequence_tensor = torch.tensor(seq_neg)
    # Make predictions on subsequences of length 400
    predictions_neg = []
    input_len = 400
    n_predictions = 5
    input_tensor = torch.zeros((1, 1, 400)).to(device)
    for i in tqdm(range(n_predictions)):
        # Could also batch these to speed up predictions
        input_tensor[:, :, :] = sequence_tensor[i:i+input_len]
        prediction = model(input_tensor).sigmoid().item()
        predictions_neg.append(prediction)
    
    print("pos signals:", seq_pos[:10])
    sequence_tensor = torch.tensor(seq_pos)
    # Make predictions on subsequences of length 400
    predictions_pos = []
    input_len = 400
    n_predictions = 5
    input_tensor = torch.zeros((1, 1, 400)).to(device)
    for i in tqdm(range(n_predictions)):
        # Could also batch these to speed up predictions
        input_tensor[:, :, :] = sequence_tensor[i:i+input_len]
        prediction = model(input_tensor).sigmoid().item()
        predictions_pos.append(prediction)
        
    predict.append((predictions_neg, predictions_pos))
    sns.kdeplot(predictions_neg, label= "neg control")
    sns.kdeplot(predictions_pos, label= "pos control")
    plt.legend()
    plt.savefig(outfig)
    plt.show()
    outfig = '../results/figures/test/addseq_predicted_prob_density_' + str(count) + '.pdf'
    plt.close()
    count +=1
```
