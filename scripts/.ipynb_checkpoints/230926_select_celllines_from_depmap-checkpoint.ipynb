```r
ReadGMT <- function(gmtpath, limit = c(0, Inf)){
  c2 = readLines(gmtpath)
  c2_list = strsplit(c2, "\t")
  c2_pathway = t(matrix(unlist(lapply(c2_list, function(x) x[1:2])), nrow = 2))
  rownames(c2_pathway) = c2_pathway[,1]
  c2_list = lapply(c2_list, function(x) x[-(1:2)])
  names(c2_list) = c2_pathway[,1]
  c2_len = lapply(c2_list, length)
  gene2path = data.frame(Gene = unlist(c2_list), PathwayID = rep(names(c2_list), c2_len),
                         stringsAsFactors = FALSE)
  limit_pathways = names(c2_list)[c2_len>=limit[1] & c2_len<=limit[2]]
  gene2path = gene2path[gene2path$PathwayID%in%limit_pathways, ]
  gene2path$PathwayName = c2_pathway[gene2path$PathwayID, 2]
  return(gene2path)
}
```

```r
# Read tpm from DepMap
cell_expr<- read.table('./data/OmicsExpressionProteinCodingGenesTPMLogp1.csv', header = T, sep = ',', row.names = 1)
genes<- colnames(cell_expr)
new_colnames<- genes  %>% str_split_i(pattern = '\\.', i = 1)
colnames(cell_expr) = new_colnames
```

```r
# Read Cellline info from DrugCell
cellLine_info<- read_excel('./data/DrugCell/Cell_Lines_Details.xlsx')
colnames(cellLine_info)   = colnames(cellLine_info)  %>% str_replace_all(pattern = '[\r\n ()]', replacement = "_")
```

```r
potency_marker<- readRDS('./data/Potency_markers_PanCancer.rds')
nfkb<- ReadGMT(gmtpath = './data/BIOCARTA_NFKB_PATHWAY.v2023.1.Hs.gmt')
potency_marker$NFKB<- nfkb$Gene
```

```r
potency_marker$StemCell<- c('NANOG', 'HOXA9', 'S100A8', 'CCR6', 'BMP8B')
```

```r
# Read Drug response from DrugCell
drug_response<- read.delim('./data/DrugCell/drugcell_all.txt', header = F, sep = '\t')
```

```r
# Read meta from DepMap
model<- read.csv('./data/Model.csv', sep = ',', header = 1)
```

```r
DLBCL_celllines<- cellLine_info  %>% filter(cellLine_info$Cancer_Type___matching_TCGA_label_  == "DLBC")  %>% pull(Sample_Name)  %>%  sort()  %>% unique()
DLBCL_celllines<- DLBCL_celllines  %>% str_replace_all(pattern = '-', replacement = '')  %>%  toupper() 
length(DLBCL_celllines)
```

```r
DLBCL_ACH_id<- model %>% filter(StrippedCellLineName %in% DLBCL_celllines)  %>% pull(ModelID)
DLBCL_ACH_id<- intersect(rownames(cell_expr),DLBCL_ACH_id)
length(DLBCL_ACH_id)
```

```r
DLBCL_expr<- cell_expr[DLBCL_ACH_id,]
```

```r
color_map<- alphabet2(n = length(DLBCL_ACH_id))
names(color_map)<- DLBCL_ACH_id

plot_potency<- function(genes, marker) {

    common_genes<- intersect(new_colnames, genes)
    message('out of ', length(genes), ' genes, ', length(common_genes), " overlapped.")
    
    expr<- DLBCL_expr[,common_genes]
    mean_expr<- data.frame('cellLine' = rownames(expr), 'meanExpr' = rowSums(expr)/ncol(expr))
    
    title = paste0(marker, ' marker expression')
    p<- mean_expr %>% 
        ggplot(aes(x=reorder(cellLine, -meanExpr), y=meanExpr, fill = cellLine)) + 
        geom_bar(stat="identity") + 
        theme_bw(base_size = 14) +
        labs(x = '', y = 'log2TPM', title = title) +
        theme(axis.text.x = element_text(angle = 90), legend.position="none") +
        scale_fill_manual(values = color_map)
    
    outFile = paste0('./results/figures/230929_', marker, '_marker_expression.pdf')
    ggsave(filename = outFile, plot = p, width = 5, height = 4)
    }

i=1
for (gene in potency_marker){
    common_genes<- intersect(new_colnames, genes)
    marker = names(potency_marker)[i]
    plot_potency(genes = gene, marker = marker)
    i = i+1
}
```

```r
i=1
for (gene in potency_marker){
    
    marker = names(potency_marker[i])
    
    common_genes<- intersect(new_colnames, gene)
    expr<- DLBCL_expr[,common_genes]
    mean_expr = rowSums(expr)/ncol(expr)
    colname = paste0(marker)
    
    if (i ==1){
        df <- data.frame('cellLine' = rownames(expr))}
    df[,colname] = mean_expr
    i = i+1
}

rownames(model) = model$ModelID
DLBCL_CCLE = model[df$cellLine,'CCLEName']
colnames(drug_response) <- c('CCLE', 'compound', 'AUC')
ave_drug_response<- drug_response  %>% group_by(CCLE)  %>% summarise(aveDrugResponse = median(AUC))  %>% as.data.frame()
rownames(ave_drug_response) = ave_drug_response$CCLE
df$DrugResistance = ave_drug_response[DLBCL_CCLE, 'aveDrugResponse']
df = df[, c('cellLine', 'Pluripotent', 'Multipotent', 'Oligopotent', 'Unipotent', 'Differentiated', 'StemCell', 'NFKB', 'Normal', 'DrugResistance')]
```

```r
target_name = read.csv('data/Repurposing_Public_23Q2_Extended_Primary_Compound_List.csv')
target_name2 = read.csv('data/DrugCell/screened_compounds_rel_8.4.csv')
target_name3 = read.table('./data/DrugCell/Drug_list.tsv', header =T, sep = '\t')
drug_name = read.table('./data/DrugCell/compound_names.txt', header = T, sep = '\t', fill = TRUE, quote = "")
```

```r
target_name2  %>%  filter(TARGET_PATHWAY == 'IGF1R signaling')  %>% select(DRUG_NAME)  %>% table()
```

```r
plot_drugResponse<- function(cell_ID, prefix){
    drug_response_cellLine <- drug_response  %>% filter(CCLE == model[cell_ID, 'CCLEName'])
    cellLine = model[cell_ID,'StrippedCellLineName']
    
    drug = drug_name[match(drug_response_cellLine$compound, drug_name$SMILE),'Name']
    drug_response_cellLine = cbind(drug_response_cellLine, drug)

    target = target_name2[match(drug_response_cellLine$drug, target_name2$DRUG_NAME), c('TARGET', 'TARGET_PATHWAY')]
    drug_response_cellLine = cbind(drug_response_cellLine, target)

    drug_response_by_target<- drug_response_cellLine  %>% filter(!is.na(drug_response_cellLine$TARGET_PATHWAY))  %>% group_by(TARGET_PATHWAY)  %>% summarise(drug_sensitivity = mean(AUC))
    
    title = paste0(cellLine, "(",cell_ID ,")", " Drug Response")
    
    color = color_map[cell_ID]
    names(color) = NULL
    
    p = drug_response_by_target  %>% 
        ggplot(aes(x = reorder(TARGET_PATHWAY, drug_sensitivity) , y = drug_sensitivity)) +
        geom_bar(stat="identity",fill = color) +
        theme_classic(base_size = 14) +
        labs(x = '', y = 'AUC (Drug Resistance)', title = title) +
        theme(axis.text.x = element_text(angle = 90, size = 13),
              axis.text.y = element_text(size = 10),
              legend.position="none") +
        coord_flip()
    
    outfile = paste0('./results/figures/', prefix, cellLine, 
                     '_drug_response_by_target.pdf')
    ggsave(filename = outfile,plot = p, width = 6, height = 4)
    }
```
